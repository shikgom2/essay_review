Grad-CAM의 전체적인 구조는 다음과 같다. 오른쪽 Image Classification, Image Captioning, Visual Question Answering은 Grad-CAM에 적용 할 수 있는 다양한 vision 문제들을 설명한다.

Image Captioning은 image에 대한 설명을 만들어내는 Task이다.
Visual question answering은 VQA라고 불리며, 어떤 질문과 이미지가 주어졌을떄 이에 답을 내는 Task이다.
