시작하기에 앞서 EfficientNet의 성능을 나타낸 그래프다. EfficientNet은 이전에 다른 네트워크와 비교했을때, Top-1 accuracy가 사용된 parameters에 비해 월등이 성능이 좋은 것을 볼 수 있다. 

EffcientNet은 CNN을 효율적으로 사용 할 수 있도록 width, depth, resolution에 집중을 한다. width는 CNN의 channel에 해당한다. CNN의 parameter과 feature을 표현하는 차원의 크기를 나타낸다. depth는 network의 깊이이다. resolution은 입력값의 width와 height의 크기이다. 입력이 클 수록 정보가 많아져 성능이 향상 될  수 있지만, featuremap이 커지는 단점이 있다.

# Compound scaling

EfficientNet은 width, depth, resolution을 최적으로 조정하기 위해 세가지 요소를 동시에 고려하는 Compound scaling을 제안한다.

아래 식을 보면 depth, width, resolution은 각각 알파, 베타, 감마로 나타내고, ϕ은 모델의 크기를 조정하기 위한 계수가 된다. 각각의 비율은 아래 강조한 조건을 만족시켜야만 한다.

아래 식을 통해 3가지 요소를 각각 조정하는게 아닌 고정된 계수에 따라 조정을 해 일정한 모델의 구조가 조절될 수 있도록 할 수 있다.

EfficientNet에서는 ϕ를 1로 고정한 뒤, 최적의 알파, 베타, 감마 값을 grid search으로 찾아낸다. 본 논문에서는 알파 값은 1.2, 베타 값은 1.1, 감마 값은 1.15를 사용하였으며, 방금 구한 3개의 scaling factor는 고정한 뒤 파이를 키워주며 모델의 사이즈를 키워주고 있다.
